```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv(
    r'datasets/Info_pais.csv',
    encoding = 'ISO-8859-1',
    delimiter=';',
    decimal=','
)
```

```python
df.head()
```

```python
df_order = df.sort_values(
    'Poblacion',
    ascending= False
)
```

```python
df_order.head()
```

```python
plt.scatter(df_order['Renta per capita'],df_order['Esperanza de vida']);
```

```python
plt.scatter(df_order['Renta per capita'],df_order['Esperanza de vida'])

plt.title('Renta per cápital VS Esperanza de vida')
plt.xlabel('Renta per cápital')
plt.ylabel('Esperanza de vida');
```

```python
df_order.head()
```

```python
df_order['Poblacion_normalizada'] = df_order['Poblacion'] / max(df_order['Poblacion']);
```

```python
df_order.head()
```

```python
df_order['Poblacion_normalizada'] = df_order['Poblacion'] / (max(df_order['Poblacion'])/ 10000)
```

```python
df_order.head()
```

```python
plt.scatter(
    df_order['Renta per capita'],
    df_order['Esperanza de vida'],
    s=df_order ['Poblacion_normalizada']
)

plt.title('Renta per cápital VS Esperanza de vida')
plt.xlabel('Renta per cápital')
plt.ylabel('Esperanza de vida');
```

```python
plt.scatter(
    df_order['Renta per capita'],
    df_order['Esperanza de vida'],
    s=df_order ['Poblacion_normalizada']
)

plt.title('Renta per cápital VS Esperanza de vida')
plt.xlabel('Renta per cápital')
plt.ylabel('Esperanza de vida')
fig = plt.gcf()
fig.set_size_inches(14.5, 10);
```

```python
plt.scatter(
    df_order['Renta per capita'],
    df_order['Esperanza de vida'],
    s=df_order ['Poblacion_normalizada'],
    c=df_order['Poblacion_normalizada']
)

plt.title('Renta per cápital VS Esperanza de vida')
plt.xlabel('Renta per cápital')
plt.ylabel('Esperanza de vida')
fig = plt.gcf()
fig.set_size_inches(14.5, 10);
```

```python
plt.scatter(
    df_order['Renta per capita'],
    df_order['Esperanza de vida'],
    s=df_order ['Poblacion_normalizada'],
    c=df_order['Poblacion_normalizada']
)

plt.title('Renta per cápital VS Esperanza de vida')
plt.xlabel('Renta per cápital')
plt.ylabel('Esperanza de vida')
fig = plt.gcf()
fig.set_size_inches(14.5, 10);

for i in range (0, 10):
    plt.annotate(
        df_order['País'][i],
        (df_order['Renta per capita'][i], df_order['Esperanza de vida'][i])
    )
```

```python
plt.scatter(
    df_order['Renta per capita'],
    df_order['Esperanza de vida'],
    s=df_order ['Poblacion_normalizada'],
    c=df_order['Poblacion_normalizada']
)

plt.title('Renta per cápital VS Esperanza de vida')
plt.xlabel('Renta per cápital')
plt.ylabel('Esperanza de vida')

fig = plt.gcf()
fig.set_size_inches(14.5, 10)

df_order.reset_index(drop=True, inplace=True);

for i in range (0, 10):
    plt.annotate(
        df_order['País'][i],
        (df_order['Renta per capita'][i], df_order['Esperanza de vida'][i])
    )
```

```python
plt.scatter(
    df_order['Renta per capita'],
    df_order['Esperanza de vida'],
    s=df_order ['Poblacion_normalizada'],
    c=df_order['Poblacion_normalizada'],
    alpha=0.5
)

plt.title('Renta per cápital VS Esperanza de vida')
plt.xlabel('Renta per cápital')
plt.ylabel('Esperanza de vida')

fig = plt.gcf()
fig.set_size_inches(14.5, 10)

df_order.reset_index(drop=True, inplace=True);

for i in range (0, 10):
    plt.annotate(
        df_order['País'][i],
        (df_order['Renta per capita'][i], df_order['Esperanza de vida'][i])
    )
```

### CONCLUSIÓN

Tras el análisis de datos, observamos que a más renta per càpital más esperanza de vida. Esto es debido a que cuanta más riqueza tienes mayores posibilidades de obtener  bienes y servivios. A pesar de ello, en algunos países pese a tener pocos ingresos (ejemplo: China, India, Russia...) se alcanzan esperanzas de vida avanzadas, a causa de ser países muy desarrollados. Sin embargo, los países subdesarrollados cuentan con poca renta per càpital y por ello bajas esperanzas de vida.
